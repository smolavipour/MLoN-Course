```python
import pandas as pd
import numpy as np
import math
import time
```

```python
#Read the data and divide it into 10 data sets
df = pd.read_csv('mnist_train.csv',index_col=False,header=None)

N = 60000 #Total numer of data samples
m = 10 #Number of workers
#Number of data
numSamples = N//m
data = {}
X = {}
Y = {}
for idx in range(m):
    data[idx] = df[idx*numSamples:(idx+1)*numSamples]
    data[idx] = data[idx].reset_index(drop=True)
    Y[idx] = data[idx][0] #labels
    X[idx] = data[idx][data[idx].columns[1:785]]
    #print(data[idx].shape)
```

```python
#Dimension of X is (N/m)x784, Y is (N/m)x1, and w is 784x1

alpha = 0.00001 #learning rate
w = 2*np.random.random(784) - 1
epsilon = 0.0001
p = 0.5

def Algo1(w,alpha):
    while 1:
        gfsum = np.zeros(784);
        gf = np.zeros(784); 
        for idx in range(m):
            #gradient of f: -2*X.T*(Y-X*w)
            gf = -2*X[idx].T.dot(Y[idx] - X[idx].dot(w)) 
            #Adding noise based on p and R (square of sigma in normal(0,sigma))
            if np.random.random() < p:
                gf = gf + np.random.normal(0, 10000)
                #print(gf.head())
            gfsum = gfsum + gf
        wPrev = w
        w = wPrev - alpha*gfsum/N
        print(max(abs(w - wPrev)))
        if max(abs(w - wPrev)) < epsilon:
            break
        alpha = alpha*0.5 #adaptive step size
    return w

start = time.time() 
w = Algo1(w,alpha)  
end = time.time()
print('time=',end-start,' seconds')
w
```
